### **How are coefficients affected?**

```python
from sklearn.datasets import load_diabetes

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
data = load_diabetes()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['TARGET'] = data.target
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
coefs = []
r2_scores = []

for i in [0, 10, 100, 1000]:
    reg = Ridge(alpha=i)
    reg.fit(x_train, y_train)

    coefs.append(reg.coef_.tolist())
    y_pred = reg.predict(x_test)
    r2_scores.append(r2_score(y_test, y_pred))
```

```python
plt.figure(figsize=(14, 9))
plt.subplot(221)
plt.bar(data.feature_names, coefs[0])
plt.title('Alpha = 0 , r2_score = {}'.format(round(r2_scores[0], 2)))

plt.subplot(222)
plt.bar(data.feature_names, coefs[1])
plt.title('Alpha = 10, r2_score = {}'.format(round(r2_scores[1], 2)))

plt.subplot(223)
plt.bar(data.feature_names, coefs[2])
plt.title('Alpha = 100, r2_score = {}'.format(round(r2_scores[2], 2)))

plt.subplot(224)
plt.bar(data.feature_names, coefs[3])
plt.title('Alpha = 100, r2_score = {}'.format(round(r2_scores[3], 2)))

plt.show()
```

### **2. Higher Cofficients are affected more**

```python
plt.figure(figsize=(12, 6))
plt.plot(alphas, np.zeros(len(alphas)), color='black', linewidth=5)
for i in range(input_array.shape[0]):
    plt.plot(alphas, input_array[i], label=data.feature_names[i])
plt.legend()
```

### **3. Impact on Bias and Variance**

```python
m = 100
X = 5 * np.random.rand(m, 1) - 2
y = 0.7 * X ** 2 - 2 * X + 3 + np.random.randn(m, 1)

plt.scatter(X, y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X.reshape(100, 1), y.reshape(100), test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=15)

X_train = poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
from mlxtend.evaluate import bias_variance_decomp

alphas = np.linspace(0, 30, 100)

loss = []
bias = []
variance = []

for i in alphas:
    reg = Ridge(alpha=i)
    avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        reg, X_train, y_train, X_test, y_test,
        loss='mse',
        random_seed=123)
    loss.append(avg_expected_loss)
    bias.append(avg_bias)
    variance.append(avg_var)
```

```python
plt.plot(alphas, loss, label='loss')
plt.plot(alphas, bias, label='Bias')
plt.plot(alphas, variance, label='Variance')
plt.ylim(0, 5)
plt.xlabel('Alpha')
plt.legend()
```

### **4. Effect of Regularization on Loss Function**

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1,
                       n_informative=1, n_targets=1, noise=20, random_state=13)

plt.scatter(X, y)


reg = LinearRegression()
reg.fit(X, y)
print(reg.coef_)
print(reg.intercept_)
```

```python
def cal_loss(m, alpha):
    return np.sum((y - m*X.ravel() + 2.29)**2) + alpha*m*m
```

```python
def predict(m):
    return m*X - 2.29
```

```python
m = np.linspace(-45, 100, 100)
plt.figure(figsize=(4, 6))
for j in [0, 10, 20, 30, 40, 50, 100]:
    loss = []
    for i in range(m.shape[0]):
        loss_i = cal_loss(m[i], j)
        loss.append(loss_i)
    plt.plot(m, loss, label='alpha = {}'.format(j))
plt.legend()
plt.xlabel('Alpha')
plt.ylabel('Loss')
plt.show()
```
