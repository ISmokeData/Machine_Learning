```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0],test_size=0.2, random_state=2)
```

```python
print("X_train",X_train.shape)
print("y_train",y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print("Accuracy of Logistic Regression", accuracy_score(y_test, y_pred1))
print("Accuracy of Decision Tree",accuracy_score(y_test,y_pred2))
```

```python
print("Logistic Regression confusion Metrix\n")
pd.DataFrame(confusion_matrix(y_test, y_pred1),columns=list(range(0,10)))
```

```python
print("Decision Tree confusion metrix\n")
pd.DataFrame(confusion_matrix(y_test, y_pred2),columns=list(range(0,10)))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision_score(y_test,y_pred1, average='weighted')
```

```python
print("recall score",recall_score(y_test, y_pred1, average='weighted'))
print("f1 score",f1_score(y_test,y_pred1, average='weighted'))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```
